- Summary Statistics for numerical Data

```python
# Display numeriacl columns
num_col=df_sales.select_dtypes("number").columns
num_col
```

```python
print(f"The summury statistics for the numerical values is")
for col in num_col :
  mean_value=df_sales[col].mean().round(2)
  median_value=df_sales[col].median()
  mode_value=df_sales[col].mode()
  print(f"The mean valve is {mean_value},the median value is {median_value} the mode value is {mode_value}")
  print("\n")
```
